```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
from sqlalchemy import create_engine
from sqlalchemy import Column, Float, String, Integer
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship, sessionmaker

# Régréssion linéaire
import sklearn.linear_model
from sklearn.model_selection import train_test_split

# Régression linéaire univariée, multivarié
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Régression linéaire polynomiale
from sklearn.preprocessing import PolynomialFeatures
from sqlalchemy.sql.sqltypes import Float, INTEGER, Integer, SMALLINT, SmallInteger

housing = pd.read_csv(
    "/home/simplon/brief_housing/housing.csv"
)
housing.head()
```

# Object Relational Mapping Database housing 

```python
def connect_db(dico):

    str_engine = (
        dico["connector"]
        + "://"
        + dico["user"]
        + ":"
        + dico["pwd"]
        + "@"
        + dico["host"]
        + ":"
        + dico["port"]
        + "/"
        + dico["bdd"]
    )
    connect = create_engine(str_engine)
    return connect


d = {
    "connector": "postgresql",
    "user": "luca",
    "pwd": "simplon",
    "host": "localhost",
    "port": "5432",
    "bdd": "housing",
}

engine = connect_db(d)
Base = declarative_base()


class Prix_Median(Base):
    __tablename__ = "prix_median"

    id_prix_median = Column(Integer, primary_key=True)

    ocean_proximity = Column(String(30))
    latitude = Column(Float)
    longitude = Column(Float)
    housing_median_age = Column(Float)
    total_rooms = Column(Float)
    total_bedrooms = Column(Float)
    population = Column(Float)
    households = Column(SmallInteger)
    median_income = Column(Float)
    median_house_value = Column(Integer)

    def __init__(
        self,
        ocean_proximity,
        latitude,
        longitude,
        housing_median_age,
        total_rooms,
        total_bedrooms,
        population,
        households,
        median_income,
        median_house_value,
    ):
        self.ocean_proximity = ocean_proximity
        self.latitude = latitude
        self.longitude = longitude
        self.housing_median_age = housing_median_age
        self.total_rooms = total_rooms
        self.total_bedrooms = total_bedrooms
        self.population = population
        self.households = households
        self.median_income = median_income
        self.median_house_value = median_house_value


Base.metadata.create_all(engine)
Session = sessionmaker(bind=engine)
session = Session()

for i in range(housing.shape[0]):
    prix_median = Prix_Median(
        housing.iloc[i]["ocean_proximity"],
        housing.iloc[i]["latitude"],
        housing.iloc[i]["longitude"],
        housing.iloc[i]["housing_median_age"],
        housing.iloc[i]["total_rooms"],
        housing.iloc[i]["total_bedrooms"],
        housing.iloc[i]["population"],
        housing.iloc[i]["households"],
        housing.iloc[i]["median_income"],
        housing.iloc[i]["median_house_value"]
    )

    session.add(prix_median)

session.commit()
session.close()
```

# 1. Modèle Regression Linéaire

```python
# Prepare the data
X = np.c_[housing[["median_income"]]]
y = np.array(housing["median_house_value"])

X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=0)

X_valid, X_test, y_valid, y_test = train_test_split(
        X_test, y_test, test_size=0.5)

print(X.shape, y.shape)

print(X_train.shape, y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)

# Visualize the data
housing.plot(kind='scatter', x="median_income", y='median_house_value')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make a prediction for Cyprus
X_new = [[22587]]  # Cyprus' GDP per capita
print(model.predict(X_new)) # outputs [[ 5.96242338]]
```

# 2. Modèle Regression Univariée

```python
##########
# Définir les variables pour le modèle de régression linaire univarié

X = np.c_[housing[["median_income"]]]
y = np.array(housing["median_house_value"])

##########
# Split du premier jeu de données (training - entrainement) à hauteur de 60%

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

##########
# Split du deuxième jeu de données validation à hauteur de 20%

X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5)

# Select a linear model
model = make_pipeline(StandardScaler(),SGDRegressor(max_iter=800, alpha=0.0001))

# Train the model
model.fit(X_train, y_train)

# metric

metric1 = {'modele': 'modele_univarié',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }
metric1
```

# 2. Modèle Regression Multivariée

```python
X = np.c_[housing[["INLAND", "<1H OCEAN", "median_income"]]]
y = np.array(housing["median_house_value"])

# Split du premier jeu de données (training - entrainement) à hauteur de 60%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

# Split du deuxième jeu de données validation à hauteur de 20%
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5)

# Select a linear model
model = make_pipeline(StandardScaler(),SGDRegressor(max_iter=800, alpha=0.0001))
#model = sklearn.linear_model.SGDRegressor()

# Train the model
model.fit(X_train, y_train)

metric2 = {'modele': 'modele_multivariée',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }
metric2
```

# 3. Modèle Regression polynomiale

```python
X = np.c_[housing[["INLAND", "<1H OCEAN", "median_income"]]]
y = np.array(housing["median_house_value"])

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Split du premier jeu de données (training - entrainement) à hauteur de 60%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

# Split du deuxième jeu de données validation à hauteur de 20%
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.5)

model = make_pipeline(StandardScaler(), PolynomialFeatures(degree=2, include_bias=False),SGDRegressor(max_iter=900, alpha=0.0001))

model.fit(X_train, y_train)

metric_polynomial = {'modele': 'modele_ploynomial',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }
metric_polynomial
```


# 4. Modèle Regression Régularisée

## Lasso

```python
model = make_pipeline(StandardScaler(),SGDRegressor(max_iter=900,alpha=0.1,penalty = 'l1'))

model.fit(X_train, y_train)

metric_Lasso = {'modele': 'model_Lasso',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }

metric_Lasso
```

## Modèle Ridge

```python
model = make_pipeline(StandardScaler(),SGDRegressor(max_iter=900,alpha=0.1,penalty = 'l2'))

model.fit(X_train, y_train)

metric_Ridge = {'modele': 'model_Ridge',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }

metric_Ridge
```

## Modèle Elasticnet

```python
model = make_pipeline(StandardScaler(),SGDRegressor(max_iter=900,alpha=0.01,penalty = 'elasticnet',l1_ratio=0.3))

model.fit(X_train, y_train)

metric_ElasticNet = {'modele': 'model_ElasticNet',
          'mean_absolute_error': mean_absolute_error(y_valid, model.predict(X_valid)),
          'mean_squared_error': mean_squared_error(y_valid, model.predict(X_valid)),
          'r2_score':r2_score(y_valid, model.predict(X_valid))
         }

metric_ElasticNet
```
